```python
def birthday_paradox_probability(n):
    days_in_year = 365
    prob_no_shared = 1.0

    for i in range(n):
        prob_no_shared *= (days_in_year - i) / days_in_year

    prob_shared = 1 - prob_no_shared
    return prob_shared
```
